# Extract strings from specific cells

```python
start_example=df.loc[0]['datastr'][45:50]
end_example=df.loc[0]['datastr'][50:55]
print(start_example)
print(end_example)
```

```python
start_vals = []
for i in range(0, len(df)): 
    start_time = df.loc[i]['datastr'][45:50]
    start_vals.append(start_time)
start_vals
```

```python
end_vals = []
for i in range(0, len(df)): 
    end_time = df.loc[i]['datastr'][50:55]
    end_vals.append(end_time)
end_vals
```

# append to the data frame

```python
df['start'] = start_vals
df
```

```python
df['end'] = end_vals
df
```

```python
df = df.drop('datastr', axis=1)
```

```python
df
```

# Save the activity duration dataset

```python
df.to_stata('India_2019_act_duration.dta') 
```
